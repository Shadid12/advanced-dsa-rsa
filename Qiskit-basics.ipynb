## Create a Quantum Circuit with One Qubit

Here's the text formatted in markdown:

In Qiskit, you can initialize a qubit using the `initialize` method. This takes a list of amplitudes as input. For example, to initialize a qubit to the state $|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$:

```python
# Import necessary libraries
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector


# Create a quantum circuit with one qubit and one classical bit
qc = QuantumCircuit(1)

# Define the state amplitudes (α, β)
alpha = 1/2**0.5  # Example amplitude for |0⟩ state (like √1/2)
beta = 1/2**0.5   # Example amplitude for |1⟩ state (like √1/2)

# Initialize the qubit to the state (α, β)
qc.initialize([alpha, beta], 0)
qc.draw('mpl')  # Display the quantum circuit
```

```python
from qiskit.visualization import plot_bloch_multivector, plot_state_city

# Step 1: Create a quantum circuit with three qubits
qc = QuantumCircuit(3)

# Step 2: Visualize the state vector to confirm initialization to |000⟩
state = Statevector.from_instruction(qc)
print(state)  # This will print the state vector representation

# Visualize the state on the Bloch sphere (for each qubit separately)
plot_bloch_multivector(state)

# Visualize the state vector using the state city plot
plot_state_city(state)
```

# Applying Hadamard Gates to a Three-Qubit System

When you apply a Hadamard gate $H$ to each qubit in the initial state $\lvert 000 \rangle$, it transforms each qubit into an equal superposition of $\lvert 0 \rangle$ and $\lvert 1 \rangle$.

The Hadamard gate acts on a single qubit as follows:

$$
H \lvert 0 \rangle = \frac{1}{\sqrt{2}} (\lvert 0 \rangle + \lvert 1 \rangle)
$$

$$
H \lvert 1 \rangle = \frac{1}{\sqrt{2}} (\lvert 0 \rangle - \lvert 1 \rangle)
$$

## Applying the Hadamard Gate to $\lvert 000 \rangle$

Since you have three qubits all in the $\lvert 0 \rangle$ state, applying the Hadamard gate to each qubit transforms the state into:

$$
\lvert \psi \rangle = H \otimes H \otimes H \lvert 000 \rangle = \left(\frac{1}{\sqrt{2}} (\lvert 0 \rangle + \lvert 1 \rangle) \right) \otimes \left(\frac{1}{\sqrt{2}} (\lvert 0 \rangle + \lvert 1 \rangle) \right) \otimes \left(\frac{1}{\sqrt{2}} (\lvert 0 \rangle + \lvert 1 \rangle) \right)
$$

Expanding this tensor product, you get:

$$
\lvert \psi \rangle = \frac{1}{\sqrt{2^3}} \left( \lvert 000 \rangle + \lvert 001 \rangle + \lvert 010 \rangle + \lvert 011 \rangle + \lvert 100 \rangle + \lvert 101 \rangle + \lvert 110 \rangle + \lvert 111 \rangle \right)
$$

This state represents an equal superposition of all possible three-qubit states, each with an amplitude of $\frac{1}{\sqrt{8}}$. 

In summary, applying the Hadamard gate to each qubit in $\lvert 000 \rangle$ creates a superposition state where all 8 possible states are equally likely to be observed.

# Expanding the Tensor Product for Three-Qubit Hadamard Gates

Let's expand the tensor product step-by-step.

## Initial State

We start with the state $\lvert 000 \rangle$ and apply a Hadamard gate $H$ to each qubit. This results in:

$$
H \otimes H \otimes H \lvert 000 \rangle = \left(H \lvert 0 \rangle\right) \otimes \left(H \lvert 0 \rangle\right) \otimes \left(H \lvert 0 \rangle\right)
$$

Applying the Hadamard gate to each qubit individually:

$$
H \lvert 0 \rangle = \frac{1}{\sqrt{2}}\left(\lvert 0 \rangle + \lvert 1 \rangle\right)
$$

Therefore:

$$
H \otimes H \otimes H \lvert 000 \rangle = \left[\frac{1}{\sqrt{2}}\left(\lvert 0 \rangle + \lvert 1 \rangle\right)\right] \otimes \left[\frac{1}{\sqrt{2}}\left(\lvert 0 \rangle + \lvert 1 \rangle\right)\right] \otimes \left[\frac{1}{\sqrt{2}}\left(\lvert 0 \rangle + \lvert 1 \rangle\right)\right]
$$

## Expanding the Tensor Product

Now, let's expand this step-by-step. We will multiply each of these terms together.

$$
\begin{aligned}
&= \frac{1}{\sqrt{2}} \left(\lvert 0 \rangle + \lvert 1 \rangle\right) \otimes \frac{1}{\sqrt{2}} \left(\lvert 0 \rangle + \lvert 1 \rangle\right) \otimes \frac{1}{\sqrt{2}} \left(\lvert 0 \rangle + \lvert 1 \rangle\right) \\
&= \frac{1}{\sqrt{2^3}} \left(\lvert 0 \rangle + \lvert 1 \rangle\right) \otimes \left(\lvert 0 \rangle + \lvert 1 \rangle\right) \otimes \left(\lvert 0 \rangle + \lvert 1 \rangle\right)
\end{aligned}
$$

Next, we expand this as a three-fold tensor product:

1. **First Tensor Product:**

   $$
   \left(\lvert 0 \rangle + \lvert 1 \rangle\right) \otimes \left(\lvert 0 \rangle + \lvert 1 \rangle\right) = \lvert 00 \rangle + \lvert 01 \rangle + \lvert 10 \rangle + \lvert 11 \rangle
   $$

2. **Second Tensor Product with the Third Qubit:**

   Now, we take the result of the first tensor product and expand it with the last qubit term:

   $$
   \left(\lvert 00 \rangle + \lvert 01 \rangle + \lvert 10 \rangle + \lvert 11 \rangle\right) \otimes \left(\lvert 0 \rangle + \lvert 1 \rangle\right)
   $$

   Distributing the terms gives:

   $$
   = \lvert 000 \rangle + \lvert 001 \rangle + \lvert 010 \rangle + \lvert 011 \rangle + \lvert 100 \rangle + \lvert 101 \rangle + \lvert 110 \rangle + \lvert 111 \rangle
   $$

## Final Result

So, the expanded state after applying a Hadamard gate to each qubit is:

$$
\lvert \psi \rangle = \frac{1}{\sqrt{8}}\left(\lvert 000 \rangle + \lvert 001 \rangle + \lvert 010 \rangle + \lvert 011 \rangle + \lvert 100 \rangle + \lvert 101 \rangle + \lvert 110 \rangle + \lvert 111 \rangle\right)
$$

This is an equal superposition of all possible three-qubit states, each with an amplitude of $\frac{1}{\sqrt{8}}$.
